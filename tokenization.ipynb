```python
import nltk
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer

print("✅ All libraries imported successfully!")

# Ejemplo simple con nltk y pandas
nltk.download('punkt')
text = "This is a test sentence for Jeisson Morales."
tokens = nltk.word_tokenize(text)

df = pd.DataFrame(tokens, columns=["Tokens"])
print(df.head())

# Visualización sencilla
sns.countplot(x="Tokens", data=df)
plt.show()
```
